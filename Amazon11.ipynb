```python
def remove_punct(desc):
    text_nopunct = "".join([char for char in desc if char not in string.punctuation])  
    return text_nopunct 

df1['descpunc'] = df1['desc'].apply(lambda x: remove_punct(x)) 

df1.head()
```

```python
import re 
def tokenize(desc):  
    tokens = re.split('\W+', desc) 
    return tokens 

df1['desctoken'] = df1['descpunc'].apply(lambda x: tokenize(x.lower ())) 

df1.head()
```

```python
def remove_stopwords(tokenized_list):  
    text = [word for word in tokenized_list if word not in stopwords]   
    return text 
```
